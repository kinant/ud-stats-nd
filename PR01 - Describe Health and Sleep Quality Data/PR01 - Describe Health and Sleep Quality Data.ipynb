## Project 1: Describe Health and Sleep Quality Data

```python
# imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# load the dataset
data = pd.read_csv("data/dataset.csv")
```

```python
# Visualize Data
data.head()
```

### Part 1: Data Information and Description

```python
data.info()
```

```python
data.describe()
```

```python
for f in data.columns.tolist():
    print(f, end=', ')
```

### Part 2: Physical Activity Level

```python
# Physical Activity Level
fig, ax = plt.subplots(1, 2, figsize=(14, 5))

# Plot 1: Physical Activity Level histogram
ax[0].hist(data['Physical Activity Level'], color='skyblue', edgecolor='black')
ax[0].set_xlabel('Physical Activity Level')
ax[0].set_ylabel('Count')
ax[0].set_title('Physical Activity Level Distribution')
ax[0].grid(axis='y', alpha=0.5)
ax[0].grid(axis='y', alpha=0.5)

# Plot 2: Physical Activity Level box plot
ax[1].boxplot(data['Physical Activity Level'])
ax[1].set_xlabel('Physical Activity Level')
ax[1].set_title('Physical Activity Level Boxplot')
ax[1].grid(axis='y', alpha=0.5)

plt.tight_layout()
plt.show()
```

```python
# Describe
data[['Physical Activity Level']].describe()
```

```python
print(f"Mean = {data['Physical Activity Level'].mean():.2f}")
print(f"Median = {data['Physical Activity Level'].median()}")
print(f"Mode = {data['Physical Activity Level'].mode()[0]}")

print(f"Variance = {data['Physical Activity Level'].var():.2f}")
print(f"std = {data['Physical Activity Level'].std():.2f}")

print(f"Skew = {data['Physical Activity Level'].skew():.2f}")
```
