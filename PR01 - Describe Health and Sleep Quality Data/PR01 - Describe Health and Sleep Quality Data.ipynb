```python
# imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# load the dataset
data = pd.read_csv("data/dataset.csv")
```

```python
# Visualize Data
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
# Physical Activity Level
fig, ax = plt.subplots(1, 2, figsize=(14, 5))

# Plot 1: Physical Activity Level histogram
ax[0].hist(data['Physical Activity Level'], color='skyblue', edgecolor='black')
ax[0].set_xlabel('Physical Activity Level')
ax[0].set_ylabel('Frequency')
ax[0].set_title('Physical Activity Level')
ax[0].grid(axis='y', alpha=0.5)

# Plot 2: Physical Activity Level box plot
ax[1].boxplot(data['Physical Activity Level'])
ax[1].set_xlabel('Physical Activity Level')
ax[1].set_ylabel('Frequency')
ax[1].set_title('Physical Activity Level')
ax[1].grid(axis='y', alpha=0.5)

plt.tight_layout()
plt.show()
```

```python
# Daily Steps
fig, ax = plt.subplots(1, 2, figsize=(14, 5))

# Plot 1: Daily Steps histogram
ax[0].hist(data['Daily Steps'], color='skyblue', edgecolor='black')
ax[0].set_xlabel('Daily Steps')
ax[0].set_ylabel('Frequency')
ax[0].set_title('Daily Steps')
ax[0].grid(axis='y', alpha=0.5)

# Plot 2: Daily Steps box plot
ax[1].boxplot(data['Daily Steps'])
ax[1].set_xlabel('Daily Steps')
ax[1].set_ylabel('Frequency')
ax[1].set_title('Daily Steps')
ax[1].grid(axis='y', alpha=0.5)

plt.tight_layout()
plt.show()
```

```python
# Heart Rate
fig, ax = plt.subplots(1, 2, figsize=(14, 5))

# Plot 1: Heart Rates histogram
ax[0].hist(data['Heart Rate'], color='skyblue', edgecolor='black')
ax[0].set_xlabel('Heart Rate')
ax[0].set_ylabel('Frequency')
ax[0].set_title('Heart Rate')
ax[0].grid(axis='y', alpha=0.5)

# Plot 2: Heart Rates box plot
ax[1].boxplot(data['Heart Rate'])
ax[1].set_xlabel('Heart Rate')
ax[1].set_ylabel('Frequency')
ax[1].set_title('Heart Rate')
ax[1].grid(axis='y', alpha=0.5)

plt.tight_layout()
plt.show()
```
